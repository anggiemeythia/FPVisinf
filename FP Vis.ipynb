```python
#a. Presentase Istri yang bekerja dan tidak bekerja 
qtyC = len(dfC.index)
qtyAll = len(gakinSBY.index)
qty = qtyAll - qtyC

labels = 'istri yang tidak bekerja dan pendidikan SLTA',''
sizes = [qtyC, qty]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.show()
```
